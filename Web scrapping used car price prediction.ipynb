```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
```

```python

city=['bengaluru','chennai','new-delhi','mumbai','pune','hyderabad','kolkata','ahmedabad','jaipur','chandigarh']
brands=['hyundai','renault','mahindra','toyota','audi','honda','maruti','tata','ford','skoda']

capa = DesiredCapabilities.CHROME
capa["pageLoadStrategy"] = "eager" #since webpage has long loading time. This command will start scraping once 
driver = webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32 (2)\chromedriver.exe")

location=[]
brand=[]
year=[]
model=[]
transmission=[]
kms=[]
variant=[]
owner=[]
fuel=[]
price=[]

for j in city:
    for l in brands:
        url='https://www.cars24.com/buy-used-{}-cars-{}/'.format(l,j)

        driver.get(url)
        
    #creating the instance to scroll down the page for data to load.
        SCROLL_PAUSE_TIME = 2
        CYCLES = 20

        html = driver.find_element_by_tag_name('html')
        html.send_keys(Keys.PAGE_DOWN)  
        html.send_keys(Keys.PAGE_DOWN)  
        time.sleep(SCROLL_PAUSE_TIME)

        for k in range(CYCLES):
            html.send_keys(Keys.PAGE_UP*5)
            html.send_keys(Keys.END)
            time.sleep(SCROLL_PAUSE_TIME)
          
        for k in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
            year.append(k.text.split()[0])
                    
        for k in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
            model.append(k.text[5:])
            
        for k in driver.find_elements_by_xpath("//p[@class='cvakB']/span"):
            transmission.append(k.text)
            
        for k in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]"):
            kms.append(k.text.split()[0])
                               
        for k in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]"):
            owner.append(k.text[:1])

        for k in driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]"):
            fuel.append(k.text)
        
        for k in driver.find_elements_by_xpath("//p[@class='cvakB']"):
            variant.append(k.text.replace(' Manual','').replace(' Automatic',''))

        for k in driver.find_elements_by_xpath("//div[@class='_7udZZ']/span"):
            price.append(k.text[1:])
            location.append(j)
            brand.append(l)
                
        print(len(fuel))
        
        df=pd.DataFrame({'Brands':brand,'Model':model,'Variant':variant,'Make Year':year,'Fuel':fuel,
                     'Transmission':transmission,'Driven Kilometers':kms,'Owners':owner,
                     'Location':location,'Price':price})
        
        df.to_csv('Carsdata.csv')
```
